# 1. Split and scale

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


X, y = df_cleaned.drop(['species', 'species_name'], axis=1), df_cleaned['species_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

print(f"Mean and std for scaled X_train: {scaled_X_train.mean():.2f}, {scaled_X_train.std()}")
print(f"Mean and std for scaled X_test: {scaled_X_test.mean():.2f}, {scaled_X_test.std():.2f}")
```

# 2. Classify with logistic regression

```python
from sklearn.linear_model import LogisticRegressionCV

# model = LogisticRegressionCV() # ??????????? hur väljer man parametarar, ????????????
model = LogisticRegressionCV(penalty = 'elasticnet', l1_ratios=[0.2, 0.2, 0.2], solver='saga', max_iter=10000, cv=5) # l1-ratio som svar... hur?
model.fit(scaled_X_train, y_train)
model.coef_
```

```python
model.l1_ratio_
```

# 3. Evaluate model

```python
y_pred_prob = model.predict_proba(scaled_X_test)
y_pred_prob[:6]
```

```python
y_pred = model.predict(scaled_X_test)
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
