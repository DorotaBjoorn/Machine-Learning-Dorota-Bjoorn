## f) Remove outliers
-remove points outside of boxplots according to Tukeys rule for boxplot

```python
# reading in data again for this section
# TODO: remove when doen with ex f

df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names) 
df['species'] = iris_data.target # add target column indicating species with 0, 1, 2
df['species_name'] = [iris_data.target_names[i] for i in iris_data.target] # NOTE: intresting code; add target column indicating species with names
df = df.drop('species', axis=1)
df.head(3)
```

```python
# remove outliers for one species and one feature - OK
df_setosa = df.loc[df['species_name'] == 'setosa']

Q25 = df_setosa['sepal width (cm)'].quantile(.25)
Q75 = df_setosa['sepal width (cm)'].quantile(.75)
IQR = Q75-Q25

lower_bound = Q25 - 1.5 * IQR
higher_bound = Q75 + 1.5 * IQR

filt = (df_setosa['sepal width (cm)'] > lower_bound) & (df_setosa['sepal width (cm)'] < higher_bound)
df_setosa_cleand_from_outliers = df_setosa.loc[filt]
df_setosa_cleand_from_outliers.info()
```

```python
# .quantile seems to be applied on all numeric columns at once returning (here) 4 values, so do not need to loop through features
Q25 = df_setosa.quantile(.25, numeric_only=True)
Q75 = df_setosa.quantile(.75, numeric_only=True)
IQR = Q75-Q25
lower_bound = Q25 - 1.5 * IQR
higher_bound = Q75 + 1.5 * IQR
higher_bound
```

```python
iris_data.target_names
```

```python
filt_list = []

for species in iris_data.target_names:
    df_species = df.loc[df['species_name'] == species]
    df_species_numerical = df_species.drop("species_name", axis=1) # dropping the non-numerical target
    
    Q25 = df_species_numerical.quantile(.25)
    Q75 = df_species_numerical.quantile(.75)
    IQR = Q75-Q25
    lower_bound = Q25 - 1.5 * IQR
    higher_bound = Q75 + 1.5 * IQR

    filt = (df_species_numerical > lower_bound) & (df_species_numerical < higher_bound)
    filt_list.append(filt)

# filt_list contains [filt_1, filt_2, filt_3], where each filt is 50 rows, 4 cols
# pd.concat(filt_list) adds filt_1, filt_2, filt_3 vertically thus mask has 150 rows and 4 cols
# .all(axis=1) creates a series with True if all True in a row and False if any False in the row
mask = pd.concat(filt_list).all(axis=1)

df_cleaned = df[mask]

print(df_cleaned.head(2))
df_cleaned.info()
```

```python
df_cleaned['species_name'].value_counts()
```

```python
# filtering out rows containing False in mask and crating list of row indexes
outlier_rows = mask[mask == False].index.tolist()
print(outlier_rows)

# filtering df with outlier_rows instead
df_cleaned = df.drop(outlier_rows)
df_cleaned.info()
```

# 1. Split and scale

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


X, y = df_cleaned.drop(['species_name'], axis=1), df_cleaned['species_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)

print(f"Mean and std for scaled X_train: {scaled_X_train.mean():.2f}, {scaled_X_train.std()}")
print(f"Mean and std for scaled X_test: {scaled_X_test.mean():.2f}, {scaled_X_test.std():.2f}")
```

# 2. Classify with logistic regression

```python
from sklearn.linear_model import LogisticRegressionCV


model = LogisticRegressionCV(penalty = 'elasticnet', l1_ratios=[0.1, 0.15, 0.2, 0.25], solver='saga', max_iter=10000, cv=5, Cs = 10)
model.fit(scaled_X_train, y_train)
print(f'{model.coef_}')
```

```python
# model evaluation according to Introduction to ML book
print(f'Score train set: {model.score(scaled_X_train, y_train)}')
print(f'Score test set: {model.score(scaled_X_test, y_test)}')
```

```python
model.l1_ratio_ # best l1 scores for each class given parameters in model
```

# 3. Evaluate model

```python
y_pred_prob = model.predict_proba(scaled_X_test) # shows probablilites for each row for the 3 classes
y_pred_prob[:6]
```

```python
y_pred = model.predict(scaled_X_test)
y_pred
```

## b) confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(cm, display_labels = model.classes_) # anvnd model.classes_ to replace default 0, 1, 2 with text labels
cm_display.plot();
```

## c) classification report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
